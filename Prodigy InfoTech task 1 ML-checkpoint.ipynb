```python
from scipy import stats

z_scores = stats.zscore(traindf['SalePrice'])
threshold = 3
outliers = (np.abs(z_scores) > threshold)
traindf_no_outliers = traindf[~outliers]
```

```python
Q1 = traindf['SalePrice'].quantile(0.25)
Q3 = traindf['SalePrice'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

traindf_no_outliers = traindf[(traindf['SalePrice'] >= lower_bound) & (traindf['SalePrice'] <= upper_bound)]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=traindf_no_outliers['SalePrice'])
plt.title('SalePrice Boxplot')
plt.show()
```

```python
outliers_mask = (traindf['SalePrice'] < lower_bound) | (traindf['SalePrice'] > upper_bound)
clean_data = traindf[~outliers_mask]
num_outliers_removed = len(traindf) - len(traindf_no_outliers)
print(f'Number of outliers removed: {num_outliers_removed}')
```
